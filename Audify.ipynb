```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
import seaborn as sns

dataset=pd.read_csv('/content/IITB_Churn_Dataset.csv')
print(dataset.shape)
dataset.head(10)
```

```python
# Missing Values
dataset.isnull().sum()
```

# Numerical Features

```python
# Numerical Variables
features_num_variable=[feature for feature in dataset.columns if dataset[feature].dtype!='O']
print('Number of Numerical Variable is ',format(len(features_num_variable)))
dataset[features_num_variable].head()
```

```python
dataset.info()
```

```python
# Numerical Variables are of two types discrete and continuous
discrete_num_variable=[feature for feature in dataset[features_num_variable] if len(dataset[feature].unique())<25]
print(len(discrete_num_variable))
dataset[discrete_num_variable].head()
```

# UNIvarte analysis

```python
# check total employee range based on target column that left
sns.countplot(x='churn',data=dataset)
plt.show()
#Take-away: in the training set, less people left than didn't.it seems there is data imbalance..
```

```python
for feature in discrete_num_variable:
    
    sns.countplot(x=feature, data=dataset)
    plt.show()
```

```python
## There is a relationship between continuous variable  and Churn

continuous_feature=[feature for feature in features_num_variable if feature not in discrete_num_variable]
print("Continuous feature Count {}".format(len(continuous_feature)))
```

```python
## Lets analyse the continuous values by creating histograms to understand the distribution

for feature in continuous_feature:
    sns.distplot(dataset[feature])
    plt.show()
```

# Bivarte analysis

```python
for feature in discrete_num_variable:

 sns.catplot(x=feature, col='churn', kind='count', data=dataset)
 plt.show()

```

```python
for feature in continuous_feature:
 sns.boxplot(x='churn', y= feature, data=dataset)
 plt.show()
```

```python
for feature in continuous_feature:
    sns.violinplot(x="churn", y=feature, data=dataset,palette='rainbow')
    plt.show()
```

```python
for feature in continuous_feature:
    sns.stripplot(x='churn', y=feature, data=dataset)
    plt.show()

```

# correlation matrix

```python
## Checking for correlation
cor_mat=dataset[continuous_feature].corr()
fig = plt.figure(figsize=(15,7))
sns.heatmap(cor_mat,annot=True)
```

```python
cor_mat=dataset[discrete_num_variable].corr()
fig = plt.figure(figsize=(15,7))
sns.heatmap(cor_mat,annot=True)
```
